```python
def get_lines():
    lines = []
    root = Tk()
    root.attributes("-topmost", True)
    root.withdraw()
    for file in filedialog.askopenfilenames():
        for line in open(file, encoding="utf-8"):
            yield line
                
def convert(text,to_type):
    try:
        if to_type == "Date":
            return text[-2:]+"/"+text[4:6]+"/"+text[:4]
        elif to_type == "Int":
            return str(int(text))
        elif to_type == "Str":
            return text
        elif to_type == "Float":
            return str(float(text.replace(",","."))).replace(".",",")
    except:
        return text
```
